# Importing Library

```python
from PIL import Image, ImageChops
from matplotlib import pyplot as plt
import matplotlib.patches as mpatches
from split_image import split_image
from tabulate import tabulate
import numpy as np
import cv2
```

# Constant value

```python
# every pixel value - meter
PIXEL_RESOLUTION = 250

# area per pixel - sqaure meter
area_per_pixel = PIXEL_RESOLUTION**2  

# blue colour hsv range
lower = (90, 50, 70)
upper = (130, 255, 255)
```

# Finding the difference


```python
imgx = Image.open("Data/1973.png")
imgy = Image.open("Data/1980.png")

plt.title("Year: 1973")
plt.imshow(imgx)
plt.show()

plt.title("Year: 1980")
plt.imshow(imgy)
plt.show()

# get pixel-by-pixel difference between the two images
diff = ImageChops.difference(imgx, imgy)
plt.title("Diffrence")
plt.imshow(diff)
plt.show()
```

# River Area for 1973


```python
# Show the original image in BGR Format
img = cv2.imread("Data/1973.png")
plt.title("Year - 1973")
plt.imshow(img)
plt.show()

# convert to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# create mask for blue color in hsv
mask = cv2.inRange(hsv, lower, upper)

# count non-zero pixels in mask that is blue pixel or water area
count1 = np.count_nonzero(mask)

# Calculate area from blue pixel count
channel_area1 = count1 * area_per_pixel / 1000000.0

# Get the shape of the image
height, width, channels = img.shape

# Calculate the total number of pixels
num_pixels = height * width

# Calculate the percentage of count in num_pixels
percentage1 = (count1 / num_pixels) * 100

print("Total Area: ",((num_pixels*250*250)/1000000) , "km2")
print("River Channel Area:", round((channel_area1), 2), "km2")
print("Percentage of water covered for the given area:", round(percentage1,2), "%")
```

# Division wise area of a river

```python
# Create a numpy array of lower and upper range blue color
lower = np.array([90, 50, 70])
upper = np.array([130, 255, 255])

areaDifference = []

# Function of calculate area by year
def get_area(image_year):
    area=[]
    for i in range(slice_size):
        img = cv2.imread(f"Split/{image_year}/{image_year}_{i}.png")  
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        mask = cv2.inRange(hsv, lower, upper)
        count = cv2.countNonZero(mask)
        channel_area = count * area_per_pixel / 1000000.0
        area.append(round(channel_area,2))
    return area

# Function call and get area of every year
areaList_1973=get_area(1973)
areaList_1980=get_area(1980)
areaList_1989=get_area(1989)
areaList_2003=get_area(2003)
areaList_2010=get_area(2010)
```

# Division wise area of a river -- Table Format

```python
table = {
    "Northing(Km)": [i*10 for i in range(slice_size)],
    "1973": [i for i in areaList_1973],
    "1980": [i for i in areaList_1980],
    "1989": [i for i in areaList_1989],
    "2003": [i for i in areaList_2003],
    "2010": [i for i in areaList_2010] 
}




# ground truth values for every year river area -->  divison wise by km (16km)
original_list_1973=[25.25,23.69,28.14,29.05,29.94,32.31,31.24,25.88,30.28,31.96,23.61,19.00,21.84,16.87,21.77,28.88]
original_list_1980=[30.39,21.98,29.25,29.81,28.41,31.01,31.02,25.40,30.06,31.46,24.32,19.75,18.27,15.25,23.72,27.57]
original_list_1989=[28.93,28.64,30.73,31.64,35.51,33.49,30.22,27.90,24.96,22.85,23.79,25.12,17.25,22.85,23.84,23.11]
original_list_2003=[25.97,23.73,25.43,25.63,32.09,34.25,35.84,25.53,27.13,24.68,17.79,16.59,16.73,24.17,23.58,23.78]
original_list_2010=[33.94,31.17,30.15,33.92,33.26,34.31,35.57,26.38,24.81,23.91,17.15,16.41,16.72,26.53,27.35,22.52]



# Calculate accuracy for every year 
accuracy_1973=[round((areaList_1973[i]/original_list_1973[i]*100),2) for i in range(slice_size)]
accuracy_1980=[round((areaList_1980[i]/original_list_1980[i]*100),2) for i in range(slice_size)]
accuracy_1989=[round((areaList_1989[i]/original_list_1989[i]*100),2) for i in range(slice_size)]
accuracy_2003=[round((areaList_2003[i]/original_list_2003[i]*100),2) for i in range(slice_size)]
accuracy_2010=[round((areaList_2010[i]/original_list_2010[i]*100),2) for i in range(slice_size)]

accuracy_list=[accuracy_1973,accuracy_1980,accuracy_1989,accuracy_2003,accuracy_2010]

accuracy_avg=[]
for i in accuracy_list:
    value=round(sum(i)/len(i),2)
    if(value<100):
        accuracy_avg.append(value)

print("Total average accuracy of 1973-2010: ",round((sum(accuracy_avg)/len(accuracy_avg)),2), "%")






# print the table
print(tabulate(table,headers='keys',tablefmt='fancy_grid',showindex=range(1,len(areaList_1973)+1)))

```
